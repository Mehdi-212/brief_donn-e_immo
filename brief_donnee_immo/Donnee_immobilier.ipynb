```python
#Importation librairie
import pandas as pd
import sqlite3
```

```python
#connection
con = sqlite3.connect("immo.db") 
cur = con.cursor()
```

```python
df_finance = pd.read_excel("Valeurs_foncieres.xlsx")
df_finance
```

```python
df_pop = pd.read_excel("donnees_communes.xlsx")
df_pop
```

```python
df_geo = pd.read_excel("fr-esr-referentiel-geographique.xlsx")
df_geo
```

```python
region = df_geo[["reg_nom","reg_code"]]
region
```

```python
region.drop_duplicates().reset_index()
```

```python
region.to_sql('Region',con ,if_exists='replace', index = False)
```

```python
departement = df_geo[["dep_nom","dep_code"]]
departement
```

```python
departement.drop_duplicates(inplace=True)
departement
```

```python
departement.to_sql('Departement',con ,if_exists='replace', index = False)
```

```python
commune = df_pop[["CODDEP","CODCOM"]]
commune
```

```python
commune["id"] = commune["CODDEP"].astype(str) + commune["CODCOM"].map("{:03}".format)
commune
```

```python
#commune.drop_duplicates().reset_index()
```

```python
commune.to_sql('Commune',con ,if_exists='replace', index = False)
```

```python
vente = df_finance[["Date mutation",'Valeur fonciere']]
vente
```

```python
vente.to_sql('Vente',con ,if_exists='replace', index = False)
```

```python
bien = df_finance[['No voie', 'Type de voie', 'Voie', 'Nombre pieces principales','Type local', 'Surface Carrez du 1er lot','Surface reelle bati']]
bien 
```

```python
bien.to_sql('Bien',con ,if_exists='replace', index = False)
```
